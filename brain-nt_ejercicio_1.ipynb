```python
import pandas as pd
```

#### 1. Importar el archivo alojado en Github para su lectura

```python
url = 'https://github.com/crisda57/Brain-Network-application/raw/main/Base%20Ejercicio%20BRN%202023.csv'
df = pd.read_csv(url, delimiter=';', encoding='utf-8', thousands=',')
```

#### 2. Obtencion de la informacion del archivo, un total de 202 encuestados y 67 columnas

```python
df.info()
```

#### 3. La estructura del archivo original es la siguiente, la cual sera modificada a la estructura solicitada

```python
df.head()
```

```python
df.rename(columns={'responseID': 'response_ID'}, inplace=True) 

# llevo las 66 columnas adicionales a filas
df_melted = pd.melt(df, id_vars=['response_ID'], var_name='variable', value_name='value')

# Obtener el orden original de las variables
column_order = list(df.columns[1:])  # excluyo la columna "response_ID"

# Ordeno el DataFrame según el orden original de las variables y 'response_ID'
df_ordered = df_melted[df_melted['variable'].isin(column_order)]

# Agrego una agrupación por "response_ID"
df_grouped = df_ordered.groupby('response_ID').apply(lambda x: x.reset_index(drop=True))

# Restablezco el índice
df_grouped.reset_index(drop=True, inplace=True)
```

```python
df_grouped
```

#### 4. Por ultimo la limpieza del archivo y descarca en formato csv

```python
# elijo response_ID como indice
df_grouped.set_index('response_ID', inplace=True)

# elimino todas los strings .Banco.X de mi columna variable
df_grouped['variable'] = df_grouped['variable'].str.replace(r'\.Banco\.\d+', '', regex=True)

# elimino nulos
df_final = df_grouped.dropna(subset=['value'])
```
